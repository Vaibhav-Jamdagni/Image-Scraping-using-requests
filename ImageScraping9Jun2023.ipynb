```python
import requests
import bs4
```

```python
url = "https://www.snapdeal.com/search?keyword=mobile&santizedKeyword=&catId=&categoryId=0&suggested=false&vertical=&noOfResults=20&searchState=&clickSrc=go_header&lastKeyword=&prodCatId=&changeBackToAll=false&foundInAll=false&categoryIdSearched=&cityPageUrl=&categoryUrl=&url=&utmContent=&dealDetail=&sort=rlvncy"
```

```python
#The get method is commonly associated with the requests library, which is a popular third-party Python library used for making HTTP requests. 
#The get method in requests is used to send a GET request to the specified URL and fetch the response

response = requests.get(url)
print(type(response))
```

```python
#The response.content attribute in the requests library contains the raw byte content of the HTTP response returned by the server. 
#This attribute is useful when you need to work with binary data, such as downloading images, files, or other non-text data from the web.
rawhtml_response = response.content
parsed_response = bs4.BeautifulSoup(rawhtml_response)
```

```python
#But we can do the above step(obtaining the html response and then parsing it) simply as
soup = bs4.BeautifulSoup(response.content)
```

```python
print(soup)
```

```python
elements = soup.findAll("img",{"class":"product-image"}) 
#or element = soup.find_all("picture")
```

```python
print(elements)
```

```python
image_links = [] 
for element in elements:
    dict_attributes = element.attrs
    if 'data-src' in dict_attributes.keys():
        image_links.append(dict_attributes['data-src'])
    else:
        image_links.append(dict_attributes['src'])

print(len(image_links))
print(len(elements))
    
```

```python
import os
#The enumerate() function in Python is a built-in function that allows you to loop over an iterable (such as a list or tuple) 
#and get both the index and the value of each item in the iterable. 
#It returns an enumerate object, which is an iterator of tuples containing the index and value of each item
for i, link in enumerate(image_links):
    response_image = requests.get(link)
    
# Extracting the image filename from the URL
    filename = os.path.join("C:\\Users\\Vaibhav Jamdagni\\Downloads\\Data Science\\Image Scraped", f"image_{i + 1}.jpg") 
    with open(filename, "wb") as file:
        file.write(response_image.content)
```
